```python
train, test = train_test_split(id_flood_min_max, test_size=0.2, random_state=12345)
feat_cols = ['F_pop_atrisk', 'POPULATION_2010','F_essentialfacilities_floodplain','F_dams','F_levees']
out_col = 'flood_final_level'
train_x = train[feat_cols]
train_y = train[out_col]
test_x = test[feat_cols]
test_y = test[out_col]
```

## classification

```python
#naïve Bayes
nb = MultinomialNB()
nb.fit(train_x,train_y)
#KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_x,train_y)
#logistic regression
lr = LogisticRegression(penalty='none')
lr.fit(train_x,train_y)
#random forest
rf = RandomForestClassifier(n_estimators=10)
rf.fit(train_x,train_y)
```

```python
test
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
km.fit(train_x)
train['kmeans']=km.predict(train_x).tolist()
test['kmeans']=km.predict(test_x).tolist()


train

# from sklearn import cluster, datasets, mixture
# gm = mixture.GaussianMixture(n_components=5, covariance_type='full').fit(train_x)
# train['GM']=gm.predict(train_x).tolist()
# train


# from sklearn.cluster import DBSCAN
# train['DB']=DBSCAN(eps=0.4, min_samples=2).fit_predict(train_x)

# # train['DB']=db.predict(train_x).tolist()
# train
```

```python
train.loc[train['kmeans']==0,'kmeans_to_level']='MH'
train.loc[train['kmeans']==1,'kmeans_to_level']='L'
train.loc[train['kmeans']==2,'kmeans_to_level']='H'
train.loc[train['kmeans']==3,'kmeans_to_level']='H'
train.loc[train['kmeans']==4,'kmeans_to_level']='H'

test.loc[test['kmeans']==0,'kmeans_to_level']='MH'
test.loc[test['kmeans']==1,'kmeans_to_level']='L'
test.loc[test['kmeans']==2,'kmeans_to_level']='H'
test.loc[test['kmeans']==3,'kmeans_to_level']='H'
test.loc[test['kmeans']==4,'kmeans_to_level']='H'
```

```python
test_y.count()
test['kmeans_to_level'].count()
```

```python
acc=pd.DataFrame({'Model':['naïve Bayes','KNN','LR','RF','Kmeans'],
                  
                  'Train_ACC':[accuracy_score(train_y,nb.predict(train_x)),
                               accuracy_score(train_y,knn.predict(train_x)),
                               accuracy_score(train_y,lr.predict(train_x)),
                               accuracy_score(train_y,rf.predict(train_x)),
                               accuracy_score(train_y,train['kmeans_to_level'])
                              ],
                  
                  'Test_ACC':[accuracy_score(test_y,nb.predict(test_x)),
                               accuracy_score(test_y,knn.predict(test_x)),
                               accuracy_score(test_y,lr.predict(test_x)),
                               accuracy_score(test_y,rf.predict(test_x)),
                             accuracy_score(test_y,test['kmeans_to_level'])]
    
    
    
})

acc['diff']=np.abs(acc['Train_ACC']-acc['Test_ACC'])
acc
```

```python
labels = acc['Model'].values.tolist()
a = acc['Train_ACC'].values.tolist()
b = acc['Test_ACC'].values.tolist()

x = np.arange(len(labels))
width = 0.3 

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*2, a, width, label='Train')
rects2 = ax.bar(x - width+0.01, b, width, label='Test')

ax.set_ylabel('Accuracy', fontsize=16)
ax.set_xlabel('Models', fontsize=16)
ax.set_title('Classifier Accuracy')
ax.set_xticks(x-0.45)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score

pre=pd.DataFrame({'Model':['naïve Bayes','KNN','LR','RF','Kmeans'],
                  
                  'Precision':[precision_score(test_y,nb.predict(test_x),average='macro'),
                               precision_score(test_y,knn.predict(test_x),average='macro'),
                               precision_score(test_y,lr.predict(test_x),average='macro'),
                               precision_score(test_y,rf.predict(test_x),average='macro'),
                             precision_score(test_y,test['kmeans_to_level'],average='macro')],
                  
                  'recall':[recall_score(test_y,nb.predict(test_x),average='macro'),
                               recall_score(test_y,knn.predict(test_x),average='macro'),
                               recall_score(test_y,lr.predict(test_x),average='macro'),
                               recall_score(test_y,rf.predict(test_x),average='macro'),
                             recall_score(test_y,test['kmeans_to_level'],average='macro')],
                  
                  'f1':[f1_score(test_y,nb.predict(test_x),average='macro'),
                               f1_score(test_y,knn.predict(test_x),average='macro'),
                               f1_score(test_y,lr.predict(test_x),average='macro'),
                               f1_score(test_y,rf.predict(test_x),average='macro'),
                             f1_score(test_y,test['kmeans_to_level'],average='macro')]
                  
    
    
})

# pre['diff']=np.abs(pre['Train_ACC']-pre['Test_ACC'])
pre
```

```python
labels = pre['Model'].values.tolist()
a = pre['Precision'].values.tolist()
b = pre['recall'].values.tolist()
c = pre['f1'].values.tolist()

x = np.arange(len(labels))
width = 0.3 

fig, ax = plt.subplots()
rects1 = ax.bar(x - width-0.01, a, width, label='Precision')
rects2 = ax.bar(x , b, width, label='Recall')
rects2 = ax.bar(x + width+0.01, c, width, label='F1-score')

# ax.set_ylabel('Precision', fontsize=16)
ax.set_xlabel('Models', fontsize=16)
# ax.set_title('Classifier Precision')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
```
